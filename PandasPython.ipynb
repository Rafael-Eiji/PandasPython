# Variáveis categóricas

```python
import pandas as pd

dfv = pd.read_csv('Dados-Pesquisa.csv', sep = ',', low_memory = False)
dfv
```

```python
dfv.info()
```

```python
dfv.columns
```

```python
#Verifica quantas linhas são nulas
dfv.isna().sum()
```

```python
#apaga todas as linhas nulas
#dfv.dropna(inplace = True)
dfv.min()
```

```python
# Transformar uma coluna em fariável categórica
pd.cut(dfv['Age'], bins = [0,20,40, 60, 100], labels = ['<20', '20-40', '40-60', '>60']).value_counts()
dfv['AgeRanges'] = pd.cut(dfv['Age'], bins = [0,20,40, 60, 100], labels = ['<20', '20-40', '40-60', '>60'])
dfv
```

```python
#Verifica a quqnatidade de jobref em cada idade e converte em porcentagem
df2 = pd.crosstab(dfv.AgeRanges, 
                  dfv.JobPref).apply(lambda x :  x/x.sum(), axis = 1)
df2.plot(kind = 'bar')
```
